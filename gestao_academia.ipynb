```python
# Caminho do arquivo Excel
excel_file = "alunos_espaco30 (1).xlsx"

# Leitura da planilha
df = pd.read_excel(excel_file, engine='openpyxl')
df.head()
```

## Limpeza de Dados
Removemos linhas onde o nome do aluno está ausente.

```python
# Remover linhas sem nome
df_clean = df.dropna(subset=['Nome'])
df_clean.head()
```

## Identificação de Alunos Faltantes
Verificamos se o aluno não possui registro em nenhum dos meses.

```python
# Verificar se o aluno está ausente em todos os meses
months = ['Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
df_clean['Faltante'] = df_clean[months].isnull().all(axis=1)
df_clean[['Nome', 'Horário', 'Faltante']].head()
```

## Quantidade de Alunos por Turma
Agrupamos os dados por horário para contar os alunos.

```python
# Contar alunos por turma
alunos_por_turma = df_clean.groupby('Horário')['Nome'].count().reset_index()
alunos_por_turma.columns = ['Horário', 'Quantidade de Alunos']
alunos_por_turma
```

## Quantidade de Faltantes por Turma
Contamos os alunos faltantes por horário.

```python
# Contar faltantes por turma
faltantes_por_turma = df_clean[df_clean['Faltante']].groupby('Horário')['Nome'].count().reset_index()
faltantes_por_turma.columns = ['Horário', 'Quantidade de Faltantes']
faltantes_por_turma
```

```python
dados = {
    "nome": ["Ana", "Carlos", "João"],
    "idade": [20, 25, 22],
    "turma": ["A", "A", "B"]
}

df_clean = pd.DataFrame(dados)
```

```python
alunos_por_turma = df_clean.groupby("turma").size().reset_index(name="quantidade")
```

```python
faltantes_por_turma = pd.DataFrame({
    "turma": ["A", "B"],
    "faltantes": [1, 0]
})
```

```python
# Criar banco de dados SQLite
conn = sqlite3.connect("academia.db")

df_clean.to_sql("alunos", conn, if_exists="replace", index=False)

alunos_por_turma.to_sql(
    "estatisticas_alunos",
    conn,
    if_exists="replace",
    index=False
)

faltantes_por_turma.to_sql(
    "estatisticas_faltantes",
    conn,
    if_exists="replace",
    index=False
)

conn.commit()
conn.close()

print("Banco criado com sucesso!")
```
