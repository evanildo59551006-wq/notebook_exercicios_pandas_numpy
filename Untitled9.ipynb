```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output
import base64

# =========================
# 📊 BANCO
# =========================

try:
    df = pd.read_excel("alunos.xlsx")
except:
    df = pd.DataFrame(columns=["Nome", "Plano", "Horario", "Mes", "Pagamento", "Valor"])

# =========================
# 🎨 CORES (SÓ NO LAYOUT)
# =========================

cor_fundo = "#000000"       # preto
cor_card = "#FFFFFF"        # branco
cor_destaque = "#7FFF00"    # verde chartreuse
cor_secundaria = "#32CD32"  # verde lima
cor_texto = "#000000"       # texto PRETO

# =========================
# 🖼️ LOGO
# =========================

def carregar_logo(caminho):
    with open(caminho, "rb") as img:
        return base64.b64encode(img.read()).decode()

logo_base64 = carregar_logo("arte_camiseta_2-removebg-preview.png")

logo = widgets.HTML(f"""
<div style="background:black;padding:15px;border-radius:10px;text-align:center">
    <img src="data:image/png;base64,{logo_base64}" width="350">
</div>
""")

# =========================
# 🔐 LOGIN
# =========================

usuario = widgets.Text(description="Usuário:")
senha = widgets.Password(description="Senha:")
btn_login = widgets.Button(description="Entrar")
saida_login = widgets.Output()

USER = "admin"
PASS = "123"

def fazer_login(b):
    with saida_login:
        clear_output()
        if usuario.value == USER and senha.value == PASS:
            print("Login realizado!")
            mostrar_sistema()
        else:
            print("Usuário ou senha incorretos")

btn_login.on_click(fazer_login)

# =========================
# 🏋️ SISTEMA
# =========================

def mostrar_sistema():
    clear_output()

    nome = widgets.Text(description="Nome:")
    plano = widgets.Dropdown(
        options=["Mensal", "Trimestral", "Anual", "Gympass", "TotalPass"],
        description="Plano:"
    )

    horario = widgets.Dropdown(
        options=["06:00","07:00","08:00","09:00","10:00","17:00","18:00","19:00","20:00"],
        description="Horário:"
    )

    mes = widgets.Dropdown(
        options=["Janeiro","Fevereiro","Março","Abril","Maio","Junho",
                 "Julho","Agosto","Setembro","Outubro","Novembro","Dezembro"],
        description="Mês:"
    )

    pagamento = widgets.Dropdown(
        options=["Pago", "Pendente"],
        description="Status:"
    )

    valor = widgets.FloatText(description="Valor:")

    btn_add = widgets.Button(description="Cadastrar")
    btn_add.style.button_color = cor_secundaria

    btn_listar = widgets.Button(description="Listar")
    btn_devedores = widgets.Button(description="Devedores")

    selecionar = widgets.Dropdown(description="Aluno:")
    mes_pag = widgets.Dropdown(options=mes.options, description="Mês:")
    btn_pagar = widgets.Button(description="Marcar Pago")
    btn_pagar.style.button_color = cor_destaque

    saida = widgets.Output()

    def atualizar_lista():
        selecionar.options = df["Nome"].unique().tolist()

    def cadastrar(b):
        global df
        with saida:
            clear_output()
            novo = pd.DataFrame([{
                "Nome": nome.value,
                "Plano": plano.value,
                "Horario": horario.value,
                "Mes": mes.value,
                "Pagamento": pagamento.value,
                "Valor": valor.value
            }])
            df = pd.concat([df, novo], ignore_index=True)
            df.to_excel("alunos.xlsx", index=False)
            atualizar_lista()
            print("Cadastrado com sucesso!")

    def listar(b):
        with saida:
            clear_output()
            display(df)

    def devedores(b):
        with saida:
            clear_output()
            display(df[df["Pagamento"] == "Pendente"])

    def pagar(b):
        global df
        with saida:
            clear_output()
            filtro = (df["Nome"] == selecionar.value) & (df["Mes"] == mes_pag.value)
            df.loc[filtro, "Pagamento"] = "Pago"
            df.to_excel("alunos.xlsx", index=False)
            print("Pagamento atualizado!")

    btn_add.on_click(cadastrar)
    btn_listar.on_click(listar)
    btn_devedores.on_click(devedores)
    btn_pagar.on_click(pagar)

    atualizar_lista()

    # 🎨 CARDS BRANCOS COM TEXTO PRETO
    card_style = widgets.Layout(
        border=f"3px solid {cor_destaque}",
        padding="15px",
        width="30%"
    )

    card_cadastro = widgets.VBox([
        widgets.HTML("<h3 style='color:black'>Cadastro</h3>"),
        nome, plano, horario, mes, pagamento, valor, btn_add
    ], layout=card_style)

    card_alunos = widgets.VBox([
        widgets.HTML("<h3 style='color:black'>Alunos</h3>"),
        btn_listar, btn_devedores
    ], layout=card_style)

    card_pagamento = widgets.VBox([
        widgets.HTML("<h3 style='color:black'>Pagamentos</h3>"),
        selecionar, mes_pag, btn_pagar
    ], layout=card_style)

    layout = widgets.HBox([card_cadastro, card_alunos, card_pagamento])

    display(logo, layout, saida)

# =========================
# 🚀 INÍCIO
# =========================

display(logo, usuario, senha, btn_login, saida_login)
```
